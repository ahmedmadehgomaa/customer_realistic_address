## EDA (Exploratory Data Analysis)

```python
# build split function to extract gov from address
def gov_split(x):
    return x.split(",")[0]
```

```python
# apply gov_split to address column
df["gov"] = df["address"].apply(gov_split)
```

```python
df.sample(10)
```

```python
# build split function to extract city from address
def city_split(x):
    return x.split(",")[1]
```

```python
# apply city_split to address column
df["city"] = df["address"].apply(city_split)
```

```python
df.sample(10)
```

```python
# drop address column 
df.drop("address", axis=1, inplace=True)
```

```python
df.head()
```

```python
# get min value from budget range column 
def get_min_budget(x):
    return x[0]
df["min_budget"] = df["budget_range"].apply(get_min_budget)
```

```python
df.sample(5)
```

```python
# replace arabic values into english at columns customer type
df["customer_type"] = df["customer_type"].replace({
    "فرد":"Individual",
    "شركه":"Company"
})
```

```python
df["customer_type"].value_counts()
```

```python
# replace arabic values into english at columns lead source
df["lead_source"] = df["lead_source"].replace({
    "واتساب":"Whatsapp"
})
```

```python
df["lead_source"].value_counts()
```

```python
# replace arabic values into english at columns gov
df["gov"] = df["gov"].replace({
    "السويس":"Suez", 
    "الجيزه":"Giza",
    "اسيوط":"Assiut",
    "أسيوط":"Assiut", 
    "الجيزة":"Giza", 
    "الدقهلية":"Dakahlia",
    "الدقهليه":"Dakahlia", 
    "الإسكندرية":"Alex", 
    "الاسكندريه":"Alex", 
    "القاهرة":"Cairo", 
    "القاهره":"Cairo"
})
```

```python
df["gov"].value_counts()
```

```python
df.sample(5)
```

```python
# create new column (year)
def get_year(x):
    return x[0:4]
```

```python
get_year("2025-01-07")
```
